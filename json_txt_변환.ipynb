```python
import os
import json

def convert_to_yolo_format(json_data):
    img_width, img_height = map(int, json_data['meta']['img_resolution'].split('x'))

    yolo_data = []

    for obj in json_data['annotations']['bbox']:
        label = obj['label']

        # 타겟 레이블만 선택
        if label == '나무전체':
            class_id = 0
        elif label == '사람전체':
            class_id = 1
        #elif label == '사람전체':
        #    class_id = 2
        elif label == '집전체':
            class_id = 3
        else:
            continue

        x_center = (obj['x'] + obj['w'] / 2) / img_width
        y_center = (obj['y'] + obj['h'] / 2) / img_height
        width = obj['w'] / img_width
        height = obj['h'] / img_height

        yolo_data.append(f"{class_id} {x_center} {y_center} {width} {height}")

    return yolo_data

def process_json_files(folder_path):
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)

            with open(file_path, 'r') as file:
                json_data = json.load(file)

            yolo_data = convert_to_yolo_format(json_data)

            yolo_file_name = file_name.replace('.json', '.txt')
            yolo_file_path = os.path.join(folder_path, yolo_file_name)

            with open(yolo_file_path, 'w') as file:
                for line in yolo_data:
                    file.write(line + '\n')

# 사용 예시
folder_path = 'datasets/Training/labels/house'
process_json_files(folder_path)
```
